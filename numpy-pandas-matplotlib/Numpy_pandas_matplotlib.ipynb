```python
import time
import numpy as np

import functions_vectorised
import functions
```

```python
# генерация данных

sum_non_neg_diag_data = []
for n in [10, 100, 1000, 10000]:
    X = np.random.randint(0, 3, (n, n))
    sum_non_neg_diag_data += [X]

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
# функция для подсчета времени работы

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
# получим время работы функций

# sum_non_neg_diag

sum_non_neg_diag_time = []
sum_non_neg_diag_vectorised_time = []

for X in sum_non_neg_diag_data:
   sum_non_neg_diag_time += [count_time(functions.sum_non_neg_diag, X.tolist())]
   sum_non_neg_diag_vectorised_time += [count_time(functions_vectorised.sum_non_neg_diag, X)]

# are_multisets_equal

are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for (x, y) in are_multisets_equal_data:
   are_multisets_equal_time += [count_time(functions.are_multisets_equal, list(x), list(y))]
   are_multisets_equal_vectorised_time += [count_time(functions_vectorised.are_multisets_equal, x, y)]

# max_prod_mod_3

max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for x in max_prod_mod_3_data:
   max_prod_mod_3_time += [count_time(functions.max_prod_mod_3, list(x))]
   max_prod_mod_3_vectorised_time += [count_time(functions_vectorised.max_prod_mod_3, x)]

# convert_image

convert_image_time = []
convert_image_vectorised_time = []

for x, y in convert_image_data:
    convert_image_time += [count_time(functions.convert_image, list(x), list(y))]
    convert_image_vectorised_time += [count_time(functions_vectorised.convert_image, x, y)]

# rle_scalar

rle_scalar_time = []
rle_scalar_vectorised_time = []

for x, y in rle_scalar_data:
    rle_scalar_time += [count_time(functions.rle_scalar, list(x), list(y))]
    rle_scalar_vectorised_time += [count_time(functions_vectorised.rle_scalar, x, y)]

# cosine_distance

cosine_distance_time = []
cosine_distance_vectorised_time = []

for x, y in cosine_distance_data:
    cosine_distance_time += [count_time(functions.cosine_distance, list(x), list(y))]
    cosine_distance_vectorised_time += [count_time(functions_vectorised.cosine_distance, x, y)]
```

```python
# ВАШИ ГРАФИКИ ЗДЕСЬ
plt.title('Vectorised vs Default sum_non_neg_diag', fontsize=15)
plt.xlabel('Размер выборки', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [len(x) for x in sum_non_neg_diag_data]
plt.plot(xVals, sum_non_neg_diag_time, color='green')
plt.plot(xVals, sum_non_neg_diag_vectorised_time, color='red')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "vectorised_func"], loc ="upper left")
plt.show()

plt.title('Vectorised vs Default are_multisets_equal', fontsize=15)
plt.xlabel('Размер выборки', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [len(x) for x, _ in are_multisets_equal_data]
plt.plot(xVals, are_multisets_equal_time, color='blue')
plt.plot(xVals, are_multisets_equal_vectorised_time, color='orange')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "vectorised_func"], loc ="upper left")
plt.show()

plt.title('Vectorised vs Default max_prod_mod_3', fontsize=15)
plt.xlabel('Размер массива', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [len(x) for x in max_prod_mod_3_data]
plt.plot(xVals, are_multisets_equal_time, color='green')
plt.plot(xVals, are_multisets_equal_vectorised_time, color='red')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "max_prod_mod_3"], loc ="upper left")
plt.show()

plt.title('Vectorised vs Default convert_image', fontsize=15)
plt.xlabel('Размер результирующей выборки', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [len(x[0]) for x, _ in convert_image_data]
plt.plot(xVals, convert_image_time, color='blue')
plt.plot(xVals, convert_image_vectorised_time, color='orange')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "convert_image"], loc ="upper left")
plt.show()

plt.title('Vectorised vs Default rle_scalar', fontsize=15)
plt.xlabel('Размер массива', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [np.sum(x, axis=0)[1] for x, _ in rle_scalar_data]
plt.plot(xVals, rle_scalar_time, color='green')
plt.plot(xVals, rle_scalar_vectorised_time, color='red')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "rle_scalar"], loc ="upper left")
plt.show()

plt.title('Vectorised vs Default cosine_distance', fontsize=15)
plt.xlabel('Размер результирующей выборки', fontsize=15)
plt.ylabel('Затраченное время, с', fontsize=15)
xVals = [len(x) * len(y) for x, y in cosine_distance_data]
plt.plot(xVals, cosine_distance_time, color='blue')
plt.plot(xVals, cosine_distance_vectorised_time, color='orange')
plt.xscale('log')
plt.yscale('log')
plt.legend(["default_func", "cosine_distance"], loc ="upper left")
plt.show()
```
